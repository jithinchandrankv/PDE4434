```python
import cv2

# Initialize the camera
camera = cv2.VideoCapture(1)

# Set up the color range for the green Uno card
green_lower = (50, 162, 50)
green_upper = (50, 255, 50)

while True:
    # Capture a frame from the camera
    ret, frame = camera.read()

    # Convert the frame to the HSV color space
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Threshold the frame to only show the green Uno card
    mask = cv2.inRange(hsv, green_lower, green_upper)

    # Find contours in the thresholded image
    contours, hierarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Check if a contour was found
    if len(contours) > 0:
        # Draw a bounding box around the contour
        x, y, w, h = cv2.boundingRect(contours[0])
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Save the image to disk
        cv2.imwrite('green_uno_card.png', frame)

        # Release the camera and exit the loop
        camera.release()
        break

    # Display the frame
    cv2.imshow('Camera', frame)

    # Check for user input to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Close the window
cv2.destroyAllWindows()
```
